```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

```

```python
features = pd.read_csv("/content/drive/My Drive/KDD_competition/training_grouped.csv")
features.head()
```

```python
features.columns
```

```python
features.columns = ['unnamed_0', 'traininglabelorder_scatterplotID', 'qtde_ponto',
       'negative', 'sample', 'mean_signalx', 'sd_signalx', 'var_signalx',
       'kurtosis_signalx', 'skew_signalx', 'mean_signaly', 'sd_signaly',
       'var_signaly', 'kurtosis_signaly', 'skew_signaly', 'L', 'X', 'XY', 'Y',
       'mean_silhouette', 'traininglabelorder_score']
```

```python
labels = np.array(features.traininglabelorder_score)
features = features.drop('unnamed_0', axis=1)
features = features.drop('traininglabelorder_scatterplotID', axis=1)
features = features.drop('L', axis=1)
features = features.drop('X', axis=1)
features = features.drop('XY', axis=1) 
features = features.drop('Y', axis=1) 
features = features.drop('traininglabelorder_score', axis=1)
features = np.array(features)
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(features,
                                                                           labels,
                                                                           test_size=0.25)
```

```python
model = RandomForestRegressor(n_estimators=100)
model.fit(train_features, train_labels)
```

```python
print('Coef. de determinação: ', model.score(train_features, train_labels))
```

```python
predict = model.predict(test_features)
```

```python
rmse = sqrt(mean_squared_error(test_labels, predict))
print('Root Mean Squared Error: ', rmse)
```
